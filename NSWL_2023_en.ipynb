# <center>STATS NWSL 2023</center>

This notebook will analyse the NWSL 2023 season. Firstly the libraries of Pyhton that are going to be used will be imported. The csv file was downloaded from [Kaggle](https://www.kaggle.com/) from the user [Dakota Mills](https://www.kaggle.com/datasets/dakotamills/2023-nwsl-stats?select=2023+NWSL+Stats+by+Team.csv) who got them [National Women's Soccer League's Website](https://www.nwslsoccer.com/) and [FBref Website](https://fbref.com/en/).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import os

from mplsoccer import add_image
```

This data are the stats for each team.

```python
equipos = pd.read_csv('2023 NWSL Stats by Team.csv')
```

```python
equipos
```

As it is shown this dataset isn't in alphabetical order, so in the next few steps that is gonna be fix. At the same time the name of the team Seattle Reign FC is goinf to be change to OL Reign FC, which was the offical name of this team for the 2023 season.

```python
equipos=equipos.replace(['Seattle Reign FC'],'OL Reign FC')
```

```python
equipos
```

```python
df1=equipos.sort_values('Team')
```

```python
df1
```

The columns that have NaN values will be removed because we can analyse them.

```python
df = df1.drop(['Unnamed: 14', 'Unnamed: 15'], axis=1)
```
